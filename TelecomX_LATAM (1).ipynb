##Informe Final: Análisis de Retención y Evasión de Clientes (Churn)


🔹 **Introducción**


El objetivo principal de este análisis es identificar los factores críticos que influyen en la evasión de clientes (Churn). La pérdida de clientes no solo impacta los ingresos directos, sino que eleva los costos de adquisición de nuevos usuarios. Comprender por qué los clientes deciden cancelar sus servicios es fundamental para diseñar estrategias de retención proactivas y personalizadas.



🔹 **Limpieza y Tratamiento de Datos**


Para garantizar la calidad de los hallazgos, se realizaron los siguientes procesos:

Importación y Traducción: Se cargó el dataset y se tradujeron las columnas al español (ej. tenure a meses_permanencia, Churn a fuga) para facilitar la interpretación del equipo de negocio.

Tratamiento de Tipos: Se identificó la necesidad de convertir variables como cargo_total a formato numérico, gestionando errores de espacios vacíos.

Ingeniería de Características: Se crearon nuevas métricas como cuenta_diaria para analizar el gasto proporcional por día.

Normalización: Se aplicó un formato estándar (snake_case) a los nombres de las columnas para asegurar la compatibilidad con las librerías de visualización y modelado.

🔹 **Análisis Exploratorio de Datos (EDA)**


A través de la visualización, identificamos patrones determinantes:

*Tasa Global:* El 32.8% de la base de clientes ha cancelado el servicio, una cifra significativamente alta que requiere intervención inmediata.

*Variables Categóricas:* Los clientes con contratos "Mes a mes" muestran una tasa de fuga drásticamente superior a los contratos anuales. El género no mostró ser un factor relevante.

*Comportamiento Numérico:* Los histogramas revelaron que la fuga se concentra en los primeros 12 meses de servicio y en clientes con cargos mensuales superiores a $70.

🔹 **Conclusiones e Insights**


El "Valle de la Muerte" Inicial: Existe un riesgo crítico de abandono durante el primer año de contrato.

Sensibilidad al Precio: Los cargos mensuales elevados están directamente asociados con una mayor probabilidad de evasión.

Fidelización por Servicios: El uso de servicios adicionales como soporte técnico y seguridad en línea presenta una correlación negativa con la fuga, actuando como "anclas" de retención.

🔹 **Recomendaciones**


Basado en los datos, se sugieren las siguientes acciones estratégicas:

Migración de Contratos: Implementar incentivos o descuentos exclusivos para que los clientes en planes "Mes a mes" migren a contratos de 1 o 2 años.

Programa "Early Bird": Crear un programa de acompañamiento intensivo durante los primeros 6 meses para asegurar que el cliente perciba el valor del servicio rápidamente.

Empaquetado de Servicios: Ofrecer soporte técnico y seguridad en línea de forma bonificada o en combo para los clientes con facturas superiores a $70, con el fin de reducir su sensibilidad al precio y aumentar su lealtad.
